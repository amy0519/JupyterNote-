```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import time
```

```python
stock_2330 = pd.read_csv('2330_202007.csv',encoding='cp950',skiprows=1)
```

```python
stock_2330 = stock_2330.iloc[0:21,:]
stock_2330
```

```python
stock_2330.dropna(axis=1,inplace=True)
stock_2330
```

```python
stock_2330
```

```python
stock_2330['西元年'] = stock_2330['日期'].map(lambda x: x.split('/')[0]) 
stock_2330['月'] = stock_2330['日期'].map(lambda x: x.split('/')[1]) 
stock_2330['日'] = stock_2330['日期'].map(lambda x: x.split('/')[2]) 

stock_2330
```

```python
stock_2330['西元年'] = stock_2330['西元年'].astype(int)
```

```python
stock_2330['西元年'] = stock_2330['西元年'].apply(lambda x: x+1911)
stock_2330
```

```python
stock_2330['西元日期'] = stock_2330['西元年'].astype('str') + stock_2330['月'] + stock_2330['日']
stock_2330
```

```python
stock_2330['西元日期'] = pd.to_datetime(stock_2330['西元日期'], format = '%Y/%m/%d')
```
