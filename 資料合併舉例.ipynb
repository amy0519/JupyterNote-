```python
import pandas as pd
```

## 自訂欄位舉例

```python
df1=pd.DataFrame({'名字':['Ａmy','Alan','Bonnie','Lisa'],
                 '國文':[95,88,23,66],
                 '數學':[96,64,33,66],
                 '英语':[66,100,33,66]}
                  ,index=[0,1,2,4])
df1
```

```python
df2=pd.DataFrame({'名字':['Wen','Kobby','Jannie','Leo'],
                  '英语':[80,80,80,67],
                  '體育':[78,90,80,78],
                  '音樂':[100,92,80,83]},
                  index=[1,2,3,5])
df2
```

# 1.concat
##   axis=0(縱向) ＆ join='outer'

```python
frame = [df1,df2]
df3=pd.concat(frame,keys=['暴走大事件','NBA'],axis=0,join='outer').fillna(0)
df3
```

##   axis=0(縱向) ＆ join='inner'

```python
df4=pd.concat(frame,keys=['暴走大事件','NBA'],axis=0,join='inner').fillna(0)
df4
```

## axis=1(橫向) ＆ join='outer'

```python
df5=pd.concat(frame,keys=['暴走大事件','NBA'],axis=1,join='outer')
df5
```

## axis=1(橫向) ＆ join='inner'

```python
df6=pd.concat(frame,keys=['暴走大事件','NBA'],axis=1,join='outer')
df6
```

# 2. Append
##   縱向合併( 默認 )兩表所有人及值，英語同一列,標記缺失值為0, 重排列索引

```python
df7=df1.append(df2,ignore_index=True).fillna(0)
df7
```
