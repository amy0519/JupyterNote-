# 分析演講重點詞

## 利用jieba套件斷詞

```python
import jieba

#開啟欲分析的文字檔並讀取
with open('speech_tsai.txt' , 'r',encoding="utf-8") as f:
    speech = f.read()
```

```python
speech
```

```python
stopWords=[]
segments=[]
remainderWords=[]

```

## 讀取中文停頓詞文字檔

```python
with open('stopwords.txt', 'r', encoding='UTF-8') as file:
    for data in file.readlines():
        data = data.strip()
        
        #將讀取出的中文停頓詞加到stopWords[]中
        stopWords.append(data)
```

## 用jieba斷詞好後寫入sements[ ]

```python
segments = jieba.cut(speech, cut_all=False)
```

## 將sements[ ]沒有在stopWords[ ]中及不是\n的詞寫入remainderWords[ ]

```python
remainderWords = list(filter(lambda e: e not in stopWords and e != '\n', segments))
print(remainderWords)
```

## 將重複詞計數

```python
dic = {}
for w in remainderWords:
    if w not in dic:
        dic[w] = 1
    else:
        dic[w] = dic[w]  + 1
#dic 
```

```python
dic.items()
```

## 將重複次數最多前十顯示 

```python
swd = sorted(dic.items(),key = lambda e: e[1], reverse=True)
for rec in swd[0:10]:
    print(rec)
```
